```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img = cv2.imread('/content/gdrive/My Drive/OpenCVimg.png',0)
plt.imshow(img, interpolation = 'bicubic')
#plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```
dim = (100,100)
resized_img = cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
plt.imshow(resized_img, interpolation = 'bicubic')
plt.show()
```

```
(h, w) = resized_img.shape[:2]
center = (w / 2, h / 2)

angle180 = 180
 
scale = 1.0

M = cv2.getRotationMatrix2D(center, angle180, scale)
rotated_image = cv2.warpAffine(resized_img, M, (h, w))
plt.imshow(rotated_image, interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```
y= 25
x= 25
h= 50
w= 50
cropped_image = rotated_image[y:y+h, x:x+w]
plt.xticks([]), plt.yticks([])
plt.imshow(cropped_image, interpolation = 'bicubic')
plt.show()
```

```
cv2.imwrite('/content/gdrive/My Drive/Logo.png',cropped_image)
```
